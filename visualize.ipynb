```python
plt.rc('axes', labelsize=12)    # fontsize of the x and y labels
fig = plt.figure(figsize=(24, 10), linewidth=0)
syl = fig.supylabel("Accuracy", fontsize=36)
syl.set_x(0.005)

# plotting parameters
linewidth =   4
markersize = 24
rotation =   45
text_font =  24
loc = "lower right"
grid_color = 'gainsboro'
xlim = [-0.5, 9.5]
ylim = [73.5, 77.5]
xticks = list(range(0, 10))

plt.rcParams['legend.title_fontsize'] = 'x-large'

## 1st subplot 
ax = fig.add_subplot(121)
ax.plot(d[1], '--.', label='Nominal', linewidth=linewidth, color=colors[0], markersize=markersize)
ax.plot(d[2], '--.', label='L1', linewidth=linewidth, color=colors[1], markersize=markersize)
ax.plot(d[3], '--.', label='BN', linewidth=linewidth, color=colors[2], markersize=markersize)
ax.plot(d[4], '--.', label='L1+BN', linewidth=linewidth, color=colors[3], markersize=markersize)
ax.hlines(y=77.12, xmin=xlim[0], xmax=xlim[1], label="Baseline", linestyle="--", linewidth=linewidth, color="black")

ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xticks(xticks)
ax.set_xticklabels(text_labels, fontsize=text_font, rotation=rotation)  

ax.grid(color=grid_color) 
legend = ax.legend(loc=loc, title='Homogeneous Input', prop={'size': 28}, fontsize=28)
plt.setp(legend.get_title(), fontsize=32)

## 2nd subplot
ax2 = fig.add_subplot(122)
ax2.plot(f[1], '--.', label='Nominal', linewidth=linewidth, color=colors[0], markersize=markersize)
ax2.plot(f[2], '--.', label='L1', linewidth=linewidth, color=colors[1], markersize=markersize)
ax2.plot(f[3], '--.', label='BN', linewidth=linewidth, color=colors[2], markersize=markersize)
ax2.plot(f[4], '--.', label='L1+BN', linewidth=linewidth, color=colors[3], markersize=markersize)
ax2.hlines(y=77.12, xmin=xlim[0], xmax=xlim[1], label="Baseline", linestyle="--", linewidth=linewidth, color="black")

ax2.set_xlim(xlim)
ax2.set_ylim(ylim)
ax2.set_xticks(xticks)
ax2.set_xticklabels(text_labels, fontsize=text_font, rotation=rotation)  # https://stackabuse.com/rotate-axis-labels-in-matplotlib/

ax2.grid(color=grid_color) 
legend = ax2.legend(loc=loc, title='Fixed INT16 Input', prop={'size': 28}, fontsize=28)
plt.setp(legend.get_title(), fontsize=32)

# ax = plt.subplot(111, xlabel='x', ylabel='y', title='title')
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(32)
for item in ([ax2.title, ax2.xaxis.label, ax2.yaxis.label] +
             ax2.get_xticklabels() + ax2.get_yticklabels()):
    item.set_fontsize(32)

plt.tight_layout()
plt.savefig('figures/uniform-extended.pdf')
```
